# Legend () and Grid() function in matplotlib


```python
import matplotlib.pyplot as plt
x= [12,33,35] 
y =[45,33,36]
x1 =[33,22,35]
y1 = [12,44,55]
```

```python
plt.plot(x, y, color='g',label="first line", linewidth=5)
plt.plot(x1, yl, color= 'b',label="second line", linewidth=5)
plt.legend(loc=2)
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
plt.plot(x, y, color='g',label="first line", linewidth=5) 
plt.title("Information")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x1, yl, color='b',label="second line", linewidth=5)
plt.grid (True, color='c') 
plt.legend (loc=2)
plt.show()
```

# How to plot graph using trigonometric function

```python
import numpy as np 
import matplotlib.pyplot as plt
fig =plt.figure()
ax= plt.axes ()
x= np.linspace(0, 1000, 10)
ax.plot (x,np.sin(x));
plt.show()
```

```python
fig =plt.figure()
ax= plt.axes()
x= np.linspace(0, 1000, 10)
ax.plot(x,np.tan(x));
plt.show()
```

# TEXTPROPS() FUNCTION TO INCREASE/DECREASE TEXT SIZE


```python
explode= (0.4,0.3,0,0)
textprops = {"fontsize":20}
colors= ['c', 'y', 'r','g'] 
plt.pie (mark, explode=explode, shadow= True, labels=sub, colors =colors, autopct="%1.1f%%",radius=0.7,textprops=textprops)
plt.show()
```

```python
explode= (0.4,0.3,0,0)
textprops = {"fontsize":20}
colors= ['c', 'y', 'r','g'] 
plt.pie (mark, explode=explode, shadow= True, labels=sub, colors =colors, autopct="%1.1f%%",radius=0.7,textprops=textprops,center=(3,5))
plt.show()
```

# BASE PROPERTY

```python
explode =(0.4,0.3,0,0)
wedgeprops ={'linewidth' :0.7, 'width' :0.4, 'edgecolor':"k"}
textprops ={'fontsize':20} 
colors =['c', 'y', 'r','g']
plt.pie(mark, explode= explode, shadow= True, labels=sub,colors =colors, autopct="%1.1f%%",radius=0.7,textprops=textprops,
center=(3,5),pctdistance=0.6,wedgeprops=wedgeprops)
plt.show()
```

# START ANGLE
